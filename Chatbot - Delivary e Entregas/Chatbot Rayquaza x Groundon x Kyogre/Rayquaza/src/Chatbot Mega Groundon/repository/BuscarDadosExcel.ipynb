```python
import pandas as pd

cardapio_df = pd.read_excel('cardapio.xlsx',skiprows=2)


def get_colunas(_dataframe):
    colunas = _dataframe.columns
    return colunas

display(cardapio_df)

```

```python
colunas_df = get_colunas(cardapio_df)

colunas_dataframe = []

for i in colunas_df:
    colunas_dataframe.append(i)

colunas_dataframe
```

```python

array_table = []

for i in range(len(colunas_df)):
    tabela = cardapio_df[[colunas_df[i]]].dropna()
    array_table.append(tabela)
    if i == 2:
        break

def juntar_df():
    # Concatenar os DataFrames horizontalmente
    concatenated_df = pd.concat([array_table[0], array_table[1], array_table[2]], axis=1)

    # Exibir o DataFrame concatenado
    concatenated_df = concatenated_df.dropna()
    return concatenated_df
juntar_df()
```

```python
import pandas as pd
import os

def gerarJson(df, tabela_num):
    output_folder = 'Cardapio Json'  # Nome da pasta de saída
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    json_data = df.to_json(orient='records', force_ascii=False)
    file_name = f'tabela{tabela_num}.json'  # Nome do arquivo
    file_path = os.path.join(output_folder, file_name)  # Caminho completo do arquivo

    with open(file_path, 'w', encoding='utf-8') as file:
        file.write(json_data)

    print(f'Arquivo {file_name} salvo com sucesso!')

def criar_tabelas_dinamicas(colunas, df, tabela_num=1):
    tabela_colunas = []
    tabela_valores = []

    tabelas = []

    for coluna in colunas:
        if 'Unnamed' in coluna:
            if tabela_colunas:
                tabela = pd.DataFrame({col: val for col, val in zip(tabela_colunas, tabela_valores)})
                tabela = tabela.dropna()
                tabelas.append(tabela)
                tabela_colunas = []
                tabela_valores = []
                tabela_num += 1
            continue

        tabela_colunas.append(coluna)
        tabela_valores.append(df[coluna])

    if tabela_colunas:
        tabela = pd.DataFrame({col: val for col, val in zip(tabela_colunas, tabela_valores)})
        tabela = tabela.dropna()
        tabelas.append(tabela)

    for tabela_df in tabelas:
        display(tabela_df)
        gerarJson(tabela_df, tabela_num)
        tabela_num += 1

    return tabelas

colunas = ['Açaí e Pitaya', '300ml', '500ml', 'Unnamed: 3', 'Tapioca', 'Preço', 'Unnamed: 6']

tabelas = criar_tabelas_dinamicas(colunas_dataframe, cardapio_df)

print('\nProcesso concluído.')
```

```python
for tabela_nome, tabela_df in tabelas.items():
    print(f"{tabela_nome}:")
    display(tabela_df)
    gerarJson(tabela_df)
    print()
```
